```python
import tensorflow as tf
import numpy as np
```

```python
data  = pd.read_csv('final_project.csv')
```

```python
train_label = np.array(data.y)
```

```python
data.x29
```

```python
pd.get_dummies(data.x24)
```

```python
data = data.join(pd.get_dummies(data.x24))
data = data.join(pd.get_dummies(data.x29))
data = data.join(pd.get_dummies(data.x30))
```

```python
data.drop('x24', axis = 1, inplace = True)
data.drop('x29', axis = 1, inplace=True)
data.drop('x30', axis = 1, inplace=True)
```

```python
data.x37 = data.x37.apply(lambda x : str(x)[1:])
data.x32 = data.x32.apply(lambda x : str(x)[:-1])
```

```python
data.x37
```

```python
train_data = np.array(data.iloc[:,:14])
```

```python
data[data.iloc[:, :50].isna()]
```

```python
#tf.convert_to_tensor(train_data)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(100))
model.add(tf.keras.layers.Dense(1))
model.compile(optimizer = "rmsprop" , loss = 'binary_crossentropy' , metrics = ['binary_crossentropy'])
history = model.fit(train_data, train_label, epochs = 5, batch_size = 64)
```
